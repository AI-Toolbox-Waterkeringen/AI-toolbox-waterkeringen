```python
import geopandas as gpd
import pandas as pd
from pathlib import Path


def main():
    output_path = Path("./output")

    df_observed = gpd.read_file(output_path / "supervisely_tags_lidar.gpkg")
    df_labeled = gpd.read_file(output_path / "fix_tags_lidar.gpkg")

    observed_opschot = df_observed[df_observed.category == "opschot"].copy()
    n_detected_opschot = observed_opschot.opschot_predicted.sum()
    n_opschot = observed_opschot.shape[0]

    observed_opschot_lidar_ornan = df_observed[
        (df_observed.category == "opschot")
        & (
            (df_observed["NumberOfReturns_0.9"] > 1)
            | pd.isnull(df_observed["NumberOfReturns_0.9"])
        )
    ].copy()
    n_detected_opschot_lidar_ornan = (
        observed_opschot_lidar_ornan.opschot_predicted.sum()
    )
    n_opschot_lidar_ornan = observed_opschot_lidar_ornan.shape[0]

    observed_opschot_lidar = df_observed[
        (df_observed.category == "opschot") & (df_observed["NumberOfReturns_0.9"] > 1)
    ].copy()
    n_detected_opschot_lidar = observed_opschot_lidar.opschot_predicted.sum()
    n_opschot_lidar = observed_opschot_lidar.shape[0]

    observed_suspect_opschot_lidar = df_observed[
        (
            (df_observed.category == "opschot")
            | (df_observed.category == "twijfel opschot")
        )
        & (df_observed["NumberOfReturns_0.9"] > 1)
    ].copy()
    n_detected_suspect_opschot_lidar = (
        observed_suspect_opschot_lidar.opschot_predicted.sum()
    )
    n_suspected_opschot_lidar = observed_suspect_opschot_lidar.shape[0]

    observed_suspect_opschot_lidar_ornan = df_observed[
        (
            (df_observed.category == "opschot")
            | (df_observed.category == "twijfel opschot")
        )
        & (
            (df_observed["NumberOfReturns_0.9"] > 1)
            | pd.isnull(df_observed["NumberOfReturns_0.9"])
        )
    ].copy()
    n_detected_suspect_opschot_lidar_ornan = (
        observed_suspect_opschot_lidar_ornan.opschot_predicted.sum()
    )
    n_suspected_opschot_lidar_ornan = observed_suspect_opschot_lidar_ornan.shape[0]

    df_labeled_opschot = df_labeled[
        df_labeled.category.isin(["bush", "shrub", "tree", "shrub tree"])
    ].copy()
    df_labeled_opschot_lidar = df_labeled[
        (df_labeled["NumberOfReturns_0.9"] > 1)
        & df_labeled.category.isin(["bush", "shrub", "tree", "shrub tree"])
    ].copy()
    df_labeled_opschot_lidar_ornan = df_labeled[
        (
            (df_labeled["NumberOfReturns_0.9"] > 1)
            | pd.isnull(df_labeled["NumberOfReturns_0.9"])
        )
        & df_labeled.category.isin(["bush", "shrub", "tree", "shrub tree"])
    ].copy()
    n_labeled = df_labeled.shape[0]
    n_labeled_opschot = df_labeled_opschot.shape[0]
    n_labeled_opschot_lidar_ornan = df_labeled_opschot_lidar_ornan.shape[0]
    n_labeled_opschot_lidar = df_labeled_opschot_lidar.shape[0]

    print(
        f"{n_detected_opschot=}, {n_opschot=}, hit percentage={n_detected_opschot / n_opschot * 100:.1f}%"
    )
    print(
        f"{n_detected_opschot_lidar_ornan=}, {n_opschot_lidar_ornan=}, hit percentage={n_detected_opschot_lidar_ornan / n_opschot_lidar_ornan * 100:.1f}%"
    )
    print(
        f"{n_detected_opschot_lidar=}, {n_opschot_lidar=}, hit percentage={n_detected_opschot_lidar / n_opschot_lidar * 100:.1f}%"
    )
    print(
        f"{n_detected_suspect_opschot_lidar=}, {n_suspected_opschot_lidar=}, hit percentage={n_detected_suspect_opschot_lidar / n_suspected_opschot_lidar * 100:.1f}%"
    )
    print(
        f"{n_detected_suspect_opschot_lidar_ornan=}, {n_suspected_opschot_lidar_ornan=}, hit percentage={n_detected_suspect_opschot_lidar_ornan / n_suspected_opschot_lidar_ornan * 100:.1f}%"
    )
    print(
        f"{n_labeled=}, {n_labeled_opschot=}, {n_labeled_opschot_lidar_ornan=}, {n_labeled_opschot_lidar=}"
    )
```
